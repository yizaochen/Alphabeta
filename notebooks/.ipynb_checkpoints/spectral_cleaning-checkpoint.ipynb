```julia
using Fretem, Potential, Alphabeta, PyPlot, Printf, PhotonOperator, JLD
```

### Part 1: Get collocation points and integration kernel for whole domain

```julia
Nh = 64    # The number of Spectral element
Np = 4     # The order of polynomial which used to interpolate and integration

# Define Physical Domain
xratio = 20 # unit: angstrom
xavg = 50   # unit: angstrom

x, w, Ldx, L = getLagrange(Np, xratio/Nh)
e_norm = x[end] - x[1]
interpo_xs = x .+ x[end]

N, xref, w0, Ldx, w = get_fem_xref_weights_basis(Nh, Np, xratio, xavg);
```

### Part 2: Set $V_{eq}(x)$ and $p_{eq}(x)$

```julia
k_eq = 0.5 # unit: kcal/mol/angstrom^2
sigma_by_keq = 1 / sqrt(2 * k_eq)
V_eq = harmonic_well_k_mean(xref, k_eq, xavg)
rho_eq = get_rhoeq(V_eq, w0)
peq = rho_eq .* rho_eq
D = 4.845e9; # Diffusion coefficient, unit: angstrom^2 s-1
```

### Part 3: Get eigenvalues and eigenvectors

```julia
Nv = 72 # Number of eigenvectors
Lambdas, Qx, rho = fem_solve_eigen_by_pref(Nh, Np, xratio, xavg, peq, D, Nv);
```

### Part 4: Plot eigenvectors for $P_{\mathrm{ref}}$

```julia
nrows = 12
ncols = 6

fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(20,40))

eigv_id = 1
for row_id = 1:nrows
    for col_id = 1:ncols
        ax = axes[row_id, col_id]
        ax.plot(xref, Qx[:,eigv_id], label=eigv_id)
        eigv_id += 1
        ax.legend(frameon=false)
    end
end
plt.tight_layout()
#plt.savefig("normal_eigen_spectrum.png", dpi=200)
plt.show()
```

### Part 8: Prepare prerequisite

```julia
# Select eigenvector from eigenvector matrix
eigv_id = 1
psi_1 = Qx_EM[:,eigv_id];
```

```julia
# Add additional two more points on the tail
psi_revised = zeros(N+2)
psi_revised[1:N] = psi_1[:]
psi_revised[N+1] = psi_1[end]
psi_revised[end] = psi_1[end];
```

```julia
# Add additional two more points on the tail
xref_1 = xref[:, 1]
x_test = zeros(N+2)
x_test[1:N] = xref_1[:]
x_test[N+1] = xref_1[end] + e_norm
x_test[end] = xref_1[end] + 2 * e_norm;
```

```julia
fig, ax = plt.subplots(figsize=(4,2))
ax.plot(x_test, psi_revised)
ax.set_ylabel("\$ \\psi_1(x) \$", fontsize=12)
ax.set_xlabel("\$ x \$", fontsize=12)
plt.tight_layout()
plt.show()
```

### Part 9: First derivative

```julia
first_derivative = diff(psi_revised) ./ diff(x_test)
x_test_2 = x_test[1:N+1];
```

```julia
fig, ax = plt.subplots(figsize=(4,2))
ax.plot(x_test[1:N+1], first_derivative)
ax.set_ylabel("\$ \\psi'_1(x) \$", fontsize=12)
ax.set_xlabel("\$ x \$", fontsize=12)
plt.tight_layout()
plt.show()
```

### Part 10: Second derivative

```julia
second_derivative = diff(first_derivative) ./ diff(x_test_2);
```

```julia
fig, ax = plt.subplots(figsize=(4,2))
ax.plot(xref, second_derivative)
ax.set_ylabel("\$ \\psi''_1(x) \$", fontsize=12)
ax.set_xlabel("\$ x \$", fontsize=12)
plt.tight_layout()
plt.show()
```
