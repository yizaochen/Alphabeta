```julia
using Fretem, Potential, Alphabeta, PyPlot, Printf, PhotonOperator, JLD, LinearAlgebra, SparseArrays
```

### Part 1: Get collocation points and integration kernel for whole domain

```julia
Nh = 64    # The number of Spectral element
Np = 4     # The order of polynomial which used to interpolate and integration

# Define Physical Domain
xratio = 0.5 # unit: angstrom
xavg = 1   # unit: angstrom

x, w, Ldx, L = getLagrange(Np, xratio/Nh)
e_norm = x[end] - x[1]
interpo_xs = x .+ x[end]

N, xref, w0, Ldx, w = get_fem_xref_weights_basis(Nh, Np, xratio, xavg);
```

### Part 2: Single Well

```julia
k_eq = 247.15 # unit: kcal/mol/angstrom^2
#k_eq = 5000
sigma_by_keq = 1 / sqrt(2 * k_eq)
V_eq = harmonic_well_k_mean(xref, k_eq, xavg)
rho_eq = get_rhoeq(V_eq, w0)
pref = rho_eq .* rho_eq
pref = max.(pref, 1e-10)
D = 500 # Diffusion coefficient, unit: angstrom^2 s-1
D_array = D .* ones(size(xref));
```

```julia
Nv = 72 # Number of eigenvectors
LQ, Qx, rho = fem_solve_eigen_by_pref(Nh, Np, xratio, xavg, rho_eq .* rho_eq, D, Nv);
```

```julia
innerprod_xarray = 1:Nv
innerprod_array = zeros(Nv)
for eigv_id = 1:Nv
    innerprod_array[eigv_id] = sum(w0 .* Qx[:, eigv_id] .* Qx[:, eigv_id])
end
```

```julia
fig, axes = plt.subplots(ncols=3, nrows=1, figsize=(12, 4))

ax = axes[1]
ax.plot(xref, V_eq)
ax.set_xlabel("x ")
ax.set_ylabel("\$ V(x) \$")

ax = axes[2]
ax.plot(xref, rho_eq .* rho_eq)
ax.set_xlabel("x ")
ax.set_ylabel("\$ p_{eq}(x) \$")

ax = axes[3]
ax.scatter(innerprod_xarray, innerprod_array, s=5, color="blue")
ax.set_ylim(0.9, 1.1)
ax.set_xticks([1,6,11,16,21,26,31,36,41,46,51,56,61,66,72])
ax.set_xlabel("\$i\$, index of eigenvector")
ax.set_ylabel("\$ \\sum_{j} w(x_j)\\psi^2_i(x_j) \$")

plt.tight_layout()
#plt.savefig("single_well_1.pdf")
plt.show()
```

```julia
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(16,12))

idx_eigvector = 1
for row_id=1:3
    for col_id = 1:4
        ax = axes[row_id, col_id]
        plot_x_Qx_lambda(ax, idx_eigvector, LQ, Qx, xref)
        idx_eigvector += 1
    end
end
plt.tight_layout()
#plt.savefig("single_well_2.pdf")
plt.show()
```

### Part 4: Triple Well

```julia
function triple_well(xref)
    factor1 = 1. / 15.47
    inner_term = 13.9 * (xref .- 1)
    V = factor1 * ((inner_term .^ 6) .- (15 * (inner_term .^ 4)) .+ (53 * (inner_term .^ 2)) .+ (2 * inner_term) .- 15.)
    V = V .+ 2.9266
    return V
end
```

```julia
V_eq = triple_well(xref)
rho_eq = get_rhoeq(V_eq, w0)
peq = rho_eq .* rho_eq
D = 500;# Diffusion coefficient, unit: s-1
```

```julia
Nv = 72 # Number of eigenvectors
LQ, Qx, rho = fem_solve_eigen_by_pref(Nh, Np, xratio, xavg, rho_eq .* rho_eq, D, Nv);
```

```julia
innerprod_xarray = 1:Nv
innerprod_array = zeros(Nv)
for eigv_id = 1:Nv
    innerprod_array[eigv_id] = sum(w0 .* Qx[:, eigv_id] .* Qx[:, eigv_id])
end
```

```julia
fig, axes = plt.subplots(ncols=3, nrows=1, figsize=(12, 4))

ax = axes[1]
ax.plot(xref, V_eq)
ax.set_xlabel("x ")
ax.set_ylabel("\$ V(x) \$")
ax.set_ylim(0,8)

ax = axes[2]
ax.plot(xref, rho_eq .* rho_eq)
ax.set_xlabel("x ")
ax.set_ylabel("\$ p_{eq}(x) \$")

ax = axes[3]
ax.scatter(innerprod_xarray, innerprod_array, s=5, color="blue")
ax.set_ylim(0.9, 1.1)
ax.set_xticks([1,6,11,16,21,26,31,36,41,46,51,56,61,66,72])
ax.set_xlabel("\$i\$, index of eigenvector")
ax.set_ylabel("\$ \\sum_{j} w(x_j)\\psi^2_i(x_j) \$")

plt.tight_layout()
#plt.savefig("triple_well_1.pdf")
plt.show()
```
