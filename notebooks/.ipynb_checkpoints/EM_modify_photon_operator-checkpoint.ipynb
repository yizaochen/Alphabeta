```julia
using Fretem, Potential, Alphabeta, PyPlot, Printf, PhotonOperator, JLD
```

### Part 1: Get collocation points and integration kernel for whole domain

```julia
Nh = 64    # The number of Spectral element
Np = 4     # The order of polynomial which used to interpolate and integration

# Define Physical Domain
xratio = 20 # unit: angstrom
xavg = 50   # unit: angstrom

x, w, Ldx, L = getLagrange(Np, xratio/Nh)
e_norm = x[end] - x[1]
interpo_xs = x .+ x[end]

N, xref, w0, Ldx, w = get_fem_xref_weights_basis(Nh, Np, xratio, xavg);
```

### Part 2: Set $V_{eq}(x)$ and $p_{eq}(x)$ and $D$

```julia
k_eq = 0.5 # unit: kcal/mol/angstrom^2
sigma_by_keq = 1 / sqrt(2 * k_eq)
V_eq = harmonic_well_k_mean(xref, k_eq, xavg)
rho_eq = get_rhoeq(V_eq, w0)
peq = rho_eq .* rho_eq
D = 4.845e9; # Diffusion coefficient, unit: angstrom^2 s-1
```

### Part 3: Get eigenvalues and eigenvectors

```julia
Nv = 72 # Number of eigenvectors
Lambdas, Qx, rho = fem_solve_eigen_by_pref(Nh, Np, xratio, xavg, peq, D, Nv);
```

### Part 4: Set Gaussian $p(x) = \frac{1}{\sigma \sqrt{2\pi}}e^{-\frac{1}{2}(\frac{x-\mu}{\sigma})^2}$

```julia
k_photon = 3 # unit: kcal/mol/angstrom^2
sigma_photon = 1 / sqrt(2 * k_photon)

sigma_txt = @sprintf "Sigma is %.3f" sigma_photon
println(sigma_txt)
```

```julia
factor1 = -1 / 2
factor2  = 1 / ( sigma_photon * sqrt(2 * pi))
f_x = factor2 .* exp.(factor1 .* ((xref .- xavg) ./ sigma_photon).^2)
f_x = f_x ./ sum(f_x)
f_x = max.(f_x, 1e-10); # Photon operator
```

```julia
sum(f_x)
```

```julia
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(6, 2))

ax.plot(xref, f_x, linewidth=1)

ax.set_xlabel("x")
ax.set_ylabel("\$ f(x) \$")
title = @sprintf "\$ k_{eq} = %3.1f \$ kcal/mol/Å\$^2\$ \$ \\sigma= %.3f \$" k_photon sigma_photon

plt.show()
```

```julia
y = 45
idx = find_nearest_point(y, xref, e_norm, interpo_xs, Np)
photon_mat = get_photon_matrix_delta(y, xref, e_norm, interpo_xs, Np, w0);
```

```julia
fig, ax = plt.subplots()
im = ax.imshow(photon_mat)
ax.set_xlabel("Grid Index")
ax.set_ylabel("Grid Index")
plt.show()
```

### Part 8: Set parameters about simulations

```julia
# physical parameters
save_freq = 1e-9 # unit: s, Save trajectory per 100 ps
total_times = 10e-6 # unit: s, totol simulate 10 us, 10e-6
dt = 1e-11 # unit: s  # Integration Times 1ps
#kbT = 1
D = 4.845e9; # Diffusion coefficient, unit: angstrom^2 s-1
tau = Int(round(total_times / save_freq, digits=0))  # Number of photons
n_interval = Int(round(save_freq / dt, digits=0));

time_info = @sprintf "Save trajectory per %d timesteps. There will be %d data points." n_interval tau
println(time_info)
```

### Part 9: Initial guess $p^0_{eq}(x)$
