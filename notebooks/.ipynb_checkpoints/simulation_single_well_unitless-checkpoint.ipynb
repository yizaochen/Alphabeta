```julia
using Fretem, Potential, PyPlot, Printf, JLD
```

### Part 1: Get collocation points and integration kernel for whole domain

```julia
Nh = 64    # The number of Spectral element
Np = 4     # The order of polynomial which used to interpolate and integration

# Define Physical Domain
xratio = 0.5 # unit: angstrom
xavg = 1   # unit: angstrom

x, w, Ldx, L = getLagrange(Np, xratio/Nh)
e_norm = x[end] - x[1]
interpo_xs = x .+ x[end]

N, xref, w0, Ldx, w = get_fem_xref_weights_basis(Nh, Np, xratio, xavg);
```

### Part 2: Set $V(x)$ and $p_{eq}(x)$ for simulation

```julia
k_ref = 247.15
Vref = harmonic_well_k_mean(xref, k_ref, xavg)
rho_eq = get_rhoeq(Vref, w0);
```

### Part 3: Langenvin dynamics simulations

```julia
# physical parameters
save_freq = 1e-7 # unit: None, Save trajectory per  ps
total_times = 10e-4 # unit: None, totol simulate time
dt = 1e-9 # unit: s  # Integration Times
#kbT = 1
D = 500; # Diffusion coefficient, unit: s-1
tau = Int(round(total_times / save_freq, digits=0))  # Number of photons
n_interval = Int(round(save_freq / dt, digits=0));

time_info = @sprintf "Save trajectory per %d timesteps. There will be %d data points." n_interval tau
println(time_info)
```

```julia
# Initialize Variables
y_record = zeros(tau+1, 1)  # In HMM, observing variables Y(t)
t_record = zeros(tau+1, 1)

ypos = xavg # Initial Position, from mean position
t = 0

# Simulation Start
for tau_id = 1:tau
    y_record[tau_id] = ypos
    t_record[tau_id] = t
    for i = 1:n_interval
        F = force_harmonic_well_k_mean(ypos, k_eq, xavg)
        ypos = ypos + D*dt*F + (dt*2*D)^(1/2) * randn()
        t += dt   
    end
end
y_record[tau+1] = ypos
t_record[tau+1] = t;
```

### Part 4: Save trajectory or Read trajectory

```julia
# Save simulated trajectory to jld file
f_out = "../trajectory/singlewell_unitless.jld"
save(f_out, "y_record", y_record, "t_record", t_record, "xref", xref, "rho_eq", rho_eq, "D", D, "k_ref", k_ref)
```

```julia
# Read simulated trajectory from jld file
f_in = "../trajectory/singlewell_unitless.jld"
y_record = load(f_in, "y_record")
t_record = load(f_in, "t_record")
xref = load(f_in, "xref")
rho_eq = load(f_in, "rho_eq")
k_ref = load(f_in, "k_ref")
D = load(f_in, "D");
```

### Part 5: Plot

```julia
fig, axes = plt.subplots(ncols=3, nrows=1, figsize=(16,4))

ax = axes[1]
ax.plot(xref, V_eq)
ax.set_xlabel("\$x\$")
ax.set_ylabel("\$ V(x) \$")
ax.set_ylim(0, 10)
ax.set_xlim(0.5, 1.5)

ax = axes[2]
ax.plot(xref, rho_eq .* rho_eq)
ax.set_xlabel("\$x\$")
ax.set_ylabel("\$ p_{eq}(x) \$")
ax.set_xlim(0.5, 1.5)

ax = axes[3]
ax.plot(t_record, y_record)
ax.set_xlabel("Time")
ax.set_ylabel("\$x\$")
ax.set_ylim(0.5, 1.5)

plt.tight_layout()
#plt.savefig("single_well_unitless_1.pdf")
plt.show()
```
