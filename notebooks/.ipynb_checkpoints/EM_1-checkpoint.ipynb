```julia
using Fretem, Potential, Alphabeta, PyPlot, Printf, JLD
```

### Part 0: Get collocation points and integration kernel for whole domain

```julia
Nh = 64    # The number of Spectral element
Np = 4     # The order of polynomial which used to interpolate and integration

# Define Physical Domain
xratio = 10 # unit: angstrom
xavg = 50   # unit: angstrom

x, w, Ldx, L = getLagrange(Np, xratio/Nh)
e_norm = x[end] - x[1]
interpo_xs = x .+ x[end]

N, xref, w0, Ldx, w = get_fem_xref_weights_basis(Nh, Np, xratio, xavg);
```

### Part 1: Set $V(x)$ and $p_{eq}(x)$ for simulation

```julia
k_ref = 1 # unit: kcal/mol/angstrom^2
Vref = harmonic_well_k_mean(xref, k_ref, xavg)
rho_eq = get_rhoeq(Vref, w0);
```

### Part 2: Set parameters about simulations

```julia
# physical parameters
save_freq = 1e-10 # unit: s, Save trajectory per 100 ps
total_times = 5e-6 # unit: s, totol simulate 5 us, 1e-6
dt = 1e-12 # unit: s  # Integration Times 1ps
#kbT = 1
D = 6.5e6; # Diffusion coefficient, unit: angstrom^2 s-1
tau = Int(total_times / save_freq)  # Number of photons
n_interval = Int(save_freq / dt);

time_info = @sprintf "Save trajectory per %d timesteps. There will be %d data points." n_interval tau
println(time_info)
```

### Part 3: Read Trajectory

```julia
# Read simulated trajectory from jld file
f_in = "../trajectory/k_1_tc_1us_tsim_5us_dt_100ps.jld"
y_record = load(f_in, "y_record")
t_record = load(f_in, "t_record")
xref = load(f_in, "xref")
rho_eq = load(f_in, "rho_eq")
k_ref = load(f_in, "k_ref")
D = load(f_in, "D");
```

```julia
# Plot
fig, axes = plt.subplots(ncols=2, nrows=1, figsize=(12,4))

ax = axes[1]
ax.plot(xref, rho_eq .* rho_eq)
ax.set_xlabel("x (\$ \\AA \$)")
ax.set_ylabel("\$ p_{eq}(x) \$")
title = @sprintf "k=%.3f  D=%.0f \$\\AA^2~ s^{-1}\$" k_ref D
ax.set_title(title)

ax = axes[2]
ax.plot(t_record, y_record)
ax.set_xlabel("Time (s)")
ax.set_ylabel("x (\$ \\AA \$)")
ax.set_ylim(40,60)
#plt.savefig("Simu_for_EM.png", dpi=200)
plt.show()
```

### Part 3: Initial guess $p^0_{eq}(x)$

```julia
k = 0.1 # unit: kcal/mol/angstrom^2
V = harmonic_well_k_mean(xref, k, xavg)
rho = get_rhoeq(V, w0)
p = rho .* rho
Nv = 36; # Number of eigenvectors
```

### Part 4: Four times forward-backward

```julia
p_em_1 = forward_backward(Nh, Np, xratio, xavg, p, D, Nv, tau, y_record, save_freq)
p_em_1 = max.(p_em_1, 1e-10);
```

```julia
p_em_2 = forward_backward(Nh, Np, xratio, xavg, p_em_1, D, Nv, tau, y_record, save_freq)
p_em_2 = max.(p_em_2, 1e-10);
```

```julia
p_em_3 = forward_backward(Nh, Np, xratio, xavg, p_em_2, D, Nv, tau, y_record, save_freq)
p_em_3 = max.(p_em_3, 1e-10);
```

```julia
p_em_4 = forward_backward(Nh, Np, xratio, xavg, p_em_3, D, Nv, tau, y_record, save_freq)
p_em_4 = max.(p_em_4, 1e-10);
```

```julia
p_em_5 = forward_backward(Nh, Np, xratio, xavg, p_em_4, D, Nv, tau, y_record, save_freq)
p_em_5 = max.(p_em_5, 1e-10);
```

```julia
p_em_6 = forward_backward(Nh, Np, xratio, xavg, p_em_5, D, Nv, tau, y_record, save_freq)
p_em_6 = max.(p_em_6, 1e-10);
```

```julia
p_em_7 = forward_backward(Nh, Np, xratio, xavg, p_em_6, D, Nv, tau, y_record, save_freq)
p_em_7 = max.(p_em_7, 1e-10);
```

```julia
p_em_8 = forward_backward(Nh, Np, xratio, xavg, p_em_7, D, Nv, tau, y_record, save_freq)
p_em_8 = max.(p_em_8, 1e-10);
```

### Part 5: Plot EM results

```julia
fig, axes = plt.subplots(ncols=4, nrows=1, figsize=(16, 4))

ax = axes[1]
ax.plot(xref, p_em_1, label="EM-1")
ax.plot(xref, rho .* rho, label="\$ p_0(x) \$")
ax.plot(xref, rho_eq .* rho_eq, label="\$ p_{ref}(x) \$")
ax.legend()
ax.set_xlabel("x position")
ax.set_ylabel("\$ p_{eq}(x) \$")

ax = axes[2]
ax.plot(xref, p_em_2, label="EM-2")
ax.plot(xref, rho .* rho, label="\$ p_0(x) \$")
ax.plot(xref, rho_eq .* rho_eq, label="\$ p_{ref}(x) \$")
ax.legend()
ax.set_xlabel("x position")
ax.set_ylabel("\$ p_{eq}(x) \$")

ax = axes[3]
ax.plot(xref, p_em_3, label="EM-3")
ax.plot(xref, rho .* rho, label="\$ p_0(x) \$")
ax.plot(xref, rho_eq .* rho_eq, label="\$ p_{ref}(x) \$")
ax.legend()
ax.set_xlabel("x position")
ax.set_ylabel("\$ p_{eq}(x) \$")

ax = axes[4]
ax.plot(xref, p_em_4, label="EM-4")
ax.plot(xref, rho .* rho, label="\$ p_0(x) \$")
ax.plot(xref, rho_eq .* rho_eq, label="\$ p_{ref}(x) \$")
ax.legend()
ax.set_xlabel("x position")
ax.set_ylabel("\$ p_{eq}(x) \$")

plt.tight_layout()
#plt.savefig("EM_test_p0_wide_1to4.png", dpi=150)
plt.show()
```
