```julia
using Fretem, Potential, Alphabeta, PyPlot, Printf, LinearAlgebra
```

### Part 1: Get collocation points and integration kernel for whole domain

```julia
Nh = 64    # The number of Spectral element
Np = 4     # The order of polynomial which used to interpolate and integration

# Define Physical Domain
xratio = 20 # unit: angstrom
xavg = 50   # unit: angstrom

x, w, Ldx, L = getLagrange(Np, xratio/Nh)
e_norm = x[end] - x[1]
interpo_xs = x .+ x[end]

N, xref, w0, Ldx, w = get_fem_xref_weights_basis(Nh, Np, xratio, xavg);
```

### Part 2: Set $V(x)$ and $p_{eq}(x)$ and $D$

```julia
k_ref = 0.5 # unit: kcal/mol/angstrom^2
Vref = harmonic_well_k_mean(xref, k_ref, xavg)
rho_eq = get_rhoeq(Vref, w0)
peq = rho_eq .* rho_eq
D = 4.845e9; # Diffusion coefficient, unit: angstrom^2 s-1
```

```julia
fig, axes = plt.subplots(ncols=2, nrows=1, figsize=(8,2))

ax = axes[1]
ax.plot(xref, Vref)
ax.set_xlabel("x")
ax.set_ylabel("V(x)")

ax = axes[2]
ax.plot(xref, peq)
ax.set_xlabel("x")
ax.set_ylabel("\$ p_{eq}(x)\$")

plt.tight_layout()
plt.show()
```

### Part 3: Get eigenvalues and eigenvectors

```julia
Nv = 36 # Number of eigenvectors
Eigenvalues, Qx, rho = fem_solve_eigen_by_pref(Nh, Np, xratio, xavg, peq, D, Nv)
save_freq = 1e-9 # unit: s, Save trajectory per 100 ps
expLQDT = exp.(-Eigenvalues .* save_freq);
```

### Part 4: Set plot functions

```julia
function plot_alphax_alpha(xref, Nv, alpha_x, alpha)
    fig, axes = plt.subplots(ncols=2, nrows=1, figsize=(8, 4))
    
    ax = axes[1]
    ax.plot(xref, alpha_x)
    ax.set_xlabel("x")
    ax.set_ylabel("\$ \\left< \\alpha_{t_0} | x \\right> \$")
    
    ax = axes[2]
    w = 0.5
    xarray = 1:Nv
    ax.bar(xarray, alpha, w)
    ax.set_xlabel("The index of eigenvector, i")
    ax.set_ylabel("\$ \\left< \\alpha_{t_0} | \\psi_{i} \\right> \$")
    ax.set_ylim(-1.1, 1.1)
    ax.axhline(0, color="grey", alpha=0.6)
    ax.set_xticks(1:3:Nv)
    return fig, axes
end

function get_alphax_by_alpha(Nv, alpha, Qx)
    N = size(Qx)[1]
    temp = zeros(N)
    for idx_eigv in 1:Nv
        temp = temp .+ (alpha[idx_eigv] .* Qx[:, idx_eigv])
    end
    return temp
end
```

### Part 5: Set $\left< \hat{\alpha}_{t_0} \right|$ where $\left< \hat{\alpha}_{t_0}|x \right> = \rho_{eq}(x)$ 

alpha_t0_x = rho_eq
alpha_t0 = Fretem.get_coefficients_by_proj(w0, alpha_t0_x, Qx, Nv);

```julia
fig, axes = plot_alphax_alpha(xref, Nv, alpha_t0_x, alpha_t0)
plt.tight_layout()
#plt.savefig("alpha_t0.png", dpi=100)
plt.show()
```

### Part 6: $\left< \hat{\alpha}_{t_0} \right| e^{-\textbf{H} \Delta t} $

```julia
alpha_t0_edt = expLQDT .* alpha_t0
alpha_t0_edt_x = get_alphax_by_alpha(Nv, alpha_t0_edt, Qx);
```

```julia
fig, axes = plot_alphax_alpha(xref, Nv, alpha_t0_edt_x, alpha_t0_edt)
axes[1].set_ylabel("\$ \\left< \\alpha_{t_0} | e^{-\\mathbf{H} \\Delta t}| x \\right> \$")
axes[2].set_ylabel("\$ \\left< \\alpha_{t_0} | e^{-\\mathbf{H} \\Delta t}| \\psi_i \\right> \$")
plt.tight_layout()
plt.show()
```

### Part 7: $\left< \hat{\alpha}_{t_1}\right|$

```julia
x_observe = 50
photon_mat = get_photon_matrix(x_observe, xref, e_norm, interpo_xs, Np, w0);
psi_photon_psi = Qx' * photon_mat * Qx
alpha_t1 =  psi_photon_psi * alpha_t0_edt
alpha_t1 = alpha_t1 / norm(alpha_t1)
alpha_t1_x = get_alphax_by_alpha(Nv, alpha_t1, Qx);
```

```julia
fig, axes = plot_alphax_alpha(xref, Nv, alpha_t1_x, alpha_t1)
axes[1].axvline(x_observe, color="red")
axes[1].set_ylabel("\$ \\left< \\hat{\\alpha}_{t_1} | x \\right> \$")
axes[2].set_ylabel("\$ \\left< \\hat{\\alpha}_{t_1}| \\psi_i \\right> \$")
plt.tight_layout()
fname = @sprintf "alpha_t1_y%d.png" x_observe
#plt.savefig(fname, dpi=100)
plt.show()
```
